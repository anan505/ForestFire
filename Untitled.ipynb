```r
library(readxl)
library(tidyverse)
library(dplyr)
library(ggplot2)
url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/00547/Algerian_forest_fires_dataset_UPDATE.csv"
```

```r
download.file(url, destfile = "data/Algerian_forest_fires_dataset_UPDATE.csv")
```

```r
Bejaia_basic <- read_csv(file = "data/Algerian_forest_fires_dataset_UPDATE.csv", skip=1, n_max=122)
# print(Bejaia_basic, n = 10)
# Bejaia_basic
SidiBelAbbes_basic <- read_csv(file = "data/Algerian_forest_fires_dataset_UPDATE.csv", skip=126)
# SidiBelAbbes_basic
# print( SidiBelAbbes_basic, n = 10)
```

```r
# change type of day and month
# add a date column to data
Bejaia_basic$year <- as.character(Bejaia_basic$year)

Bejaia_with_date<-tidyr::unite(Bejaia_basic, "date", year, month,day, remove = FALSE,sep = "-" )
# Bejaia_with_date

#FWI with color of fire
FWI_time_range <- ggplot(data=Bejaia_with_date, aes (x=date, y=FWI)) +
geom_point(aes(colour = Classes)) + 
labs(x = "month", y = "FWI with fire", colour = "Colours") +
theme(text =  element_text(size=20))
FWI_time_range


########TODO: how to change xlab
```
